```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.drop(['class', 'who', 'adult_male', 'deck', 'embark_town','alive'],axis =1 , inplace = True)
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
df.dropna(subset = ["embarked"], inplace = True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.head()
```

```python
df['sex'] = le.fit_transform(df['sex'])
df.head()
```

```python
df['embarked'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df = pd.get_dummies(df, columns=['embarked'])
```

```python
df = df.astype(int)
```

```python
df.head()
```

```python
X = df.drop('survived' ,axis =1 )
y = df['survived']
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

# K nearest neighbors 

```python
# K nearest neighbors 
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5)
```

```python
knn_model.fit(X_train_scaled,y_train)
```

```python
y_knn_pred = knn_model.predict(X_test_scaled)
```

```python
accuracy_score(y_test, y_knn_pred)
```

```python
confusion_matrix(y_test, y_knn_pred)
```

```python
print(classification_report(y_test, y_knn_pred))
```

# Naive bayes without standard scaling 

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model_NB = GaussianNB()
```

```python
model_NB.fit(X_train,y_train)
```

```python
y_NB_pred = model_NB.predict(X_test)
```

```python
y_NB_pred
```

```python
accuracy_score(y_test, y_NB_pred)
```

```python
confusion_matrix(y_test, y_NB_pred)
```

```python
print(classification_report(y_test, y_NB_pred))
```
